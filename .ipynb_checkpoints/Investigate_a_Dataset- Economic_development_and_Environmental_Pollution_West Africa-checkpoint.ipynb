## Introduction

### Dataset Description 

>This project is titled ***Economic Development and Environment Pollution in West African Countries***. The project looks at how economic development (GNI) affects the population and environment in 16 West African countries between the years 2009 to 2018. The economic indicator GNI would be used to check how economic growth affects population and environmental quality using environmental indicators such as Co2 emissions and forest area using West African countries as the case study.
.

>Data used in this project is from Gapminder which collects information about how people live their lives in different countries, tracked across the years, and on a number of different indicators. See a brief description below of each of the indicators used in this project.

>1. ***Total GNI (PPP, current international dollars $)***: Gross national income converted to international dollars using purchasing power parity rates. GNI is the total amount of money earned by a nation’s people and businesses (Investopedia, 2022). This economic indicator is used to track a nation’s wealth from year to year.
2. ***CO2 Emissions (tonnes per person)***: Carbon dioxide emissions from the burning of fossil fuels (metric tonnes of CO2 per person). CO2 emissions are the primary drivers of climate change.
3. ***Forest area (sq, km)***: Forest area is land under natural or planted stands of trees of at least 5 meters in situ. The loss of trees and vegetation can cause global warming, desertification, flooding and other environmental problems.
4. ***Population, total***: The total number of people in a country or region.

## Data Cleaning
 

### The data cleaning would involve converting letters found in the cells to values and changing the datatypes from string to float

```python
# Creating a function to search through the dataframe, and find cells with letters, 
#     extract them and map them to the actual value meant by the letters attached to the digits

def clean_df(df):
    col_list= list(df.columns)
    for col_name in col_list:
        df[col_name]= df[col_name].replace(r'[BKMbkm]+$', '', regex=True).astype(float) * df[col_name].str.extract(r'[\d\.]+([B,b,K,M,k,m]+)', expand=False).fillna(1).replace(['K','k','M','m','B','b'], [10**3, 10**3, 10**6, 10**6, 10**9, 10**9]).astype(int)
    return df
```

```python
# Applying the function to clean the forest dataframe
forest_area_clean= clean_df(forest_area_subset.copy())
forest_area_clean.head(1)
```

```python
# Applying the function to clean the population dataframe
pop_clean= clean_df(pop_subset.copy())
pop_clean.head(1)
```

```python
# Applying the function to clean the GNI dataframe
gni_clean= clean_df(gni_subset)
gni_clean.head(1)
```

```python
# checking datatypes after cleaning
print(co2_emissions_subset.info())
print(forest_area_clean.info())
print(pop_clean.info())
print(gni_clean.info())
```

## Exploratory Data Analysis


### Research Question 1: Which West African countries are doing better than others in GNI, forest preservation, and controlling Co2 emissions and population?

```python
# Lets look at the changes in each individual metric over the years, 2009 - 2018
def plot_line(df, title, label):
    """
    This functions takes in a dataframe out of the cleaned dataframes and plots a line of all columns
    """
    df_t= df.T.reset_index()
    fig, axes= plt.subplots(figsize=(10,10))
    c_list= list(df_t.columns)
    for i in range(1, len(df_t.columns)):
        sns.lineplot(x='index', y= c_list[i], data=df_t, label=c_list[i])
        plt.ylabel(label)
        plt.xlabel('Years')
        plt.title(title)
    plt.legend()
    return plt.plot()
```

```python
# Lets look at GNI for West African countries over the years, 2009 - 2018
plot_line(gni_clean, 'Changes in GNI of West African countries, 2009 - 2018.', 'GNI ($)')
```

>##### The line chart above shows GNI for each of the countries. The line chart shows that Nigeria has the highest GNI at almost 1 trillion PPP dollars, a lot higher than any of the other countries. Ghana is second with 1.6 billion PPP dollars.

```python
# Lets look at Forest Area for West African countries over the years, 2009 - 2018
plot_line(forest_area_clean, 
          'Changes in Forest area of West African countries, 2009 - 2018.', 'Forest area (sq km)')
```

>##### The line chart above shows forest area for each of the countries. The line chart shows that although Nigeria has the highest forest area, its forest area declined from roughly 280,000 sq km in 2009 to roughly 240,0000 sq km in 2018. Mali is second with a steady constant forest area throughout the years at roughly 130,0000 sq km in 2018.

```python
# Lets look at Co2 Emissions in West African countries over the years, 2009 - 2018
plot_line(co2_emissions_subset, 
          'Changes in C02 emissions of West African countries, 2009 - 2018.', 'Emission (tonnes per person)')
```

>##### The line chart above shows CO2 emissions for each of the countries. The line chart shows a steady rise in CO2 emissions in most countries from the years 2009 to 2018. Some notable exceptions are Cape Verde and Cote d'Ivoire.

```python
# Lets look at population of West African countries over the years 2009 - 2018
plot_line(pop_clean, 
          'Changes in Population of West African countries, 2009 - 2018.', 'Population')
```

>##### The line chart above shows population for each of the countries. The line chart shows that by Nigeria is by far more population than any other West African countries. There is a continued rise in all West Africa's population.

### Research question 2: Can the growth in population be attributed to GNI growth in these countries?

```python
# Creating a formula to create a 4x4 subplot comparing two metrics on per country.

def comparison_plot(df1, df2, l1, l2, title):
    """
    This function takes in 5 arguements.
    df1: first dataframe
    df2: second dataframe
    l1: label for the first dataframe as a string
    l2: label for the second dataframe as a string
    title: to title the extire subplot
    """
    df_use1= df1.T.reset_index()
    df_use2= df2.T.reset_index()
    ax_list=[]
    c_list= list(df_use1.columns)
    fig, axes= plt.subplots(4,4,figsize=(20,20))
    
    for a in range(4):
        for b in range(4):
            ax_list.append([a, b])

    for i in range(len(c_list)-1):
        axes[ax_list[i][0], ax_list[i][1]].plot(df_use1['index'], df_use1[c_list[i+1]], color='r', label= l1)
        axes[ax_list[i][0], ax_list[i][1]].set_title(c_list[i+1])
        axes[ax_list[i][0], ax_list[i][1]].legend(loc='upper left')
        axes[ax_list[i][0], ax_list[i][1]].set_xlabel('Years')
        axes[ax_list[i][0], ax_list[i][1]].tick_params(axis='x', labelrotation=45)
        axes[ax_list[i][0], ax_list[i][1]].set_ylabel(l1, color='r')
        axes2= axes[ax_list[i][0], ax_list[i][1]].twinx()
        axes2.plot(df_use2['index'], df_use2[c_list[i+1]], color= 'b', label= l2)
        axes2.set_ylabel(l2, color='b')
        axes2.legend(loc= 'upper center')
        fig.suptitle(title, y=1, fontsize= 'x-large')
        
        

    return plt.tight_layout()
```

```python
# Plotting line graphs of GNI and population for each of the countries in West Africa
comparison_plot(gni_clean, pop_clean, 'GNI', 
                'POP', 'Changes in GNI against Population emmissions of West African countries, 2009 - 2018.')
```

>##### The line charts above show the different GNI and pop in each of the countries. It clearly shows the continued increase in population over a less consistent growth in GNI in all selected countries from 2009 to 2018.

### Research Question 3: How have countries GNI affected their CO2?

```python
# Plotting line graphs to compare the relationship between GNI and CO2 over the years, 2009 - 2018
comparison_plot(gni_clean, co2_emissions_subset, 'GNI', 'C02', 
                'Changes in GNI against C02 emmissions of West African countries, 2009 - 2018.')
```

 >##### The line charts above show the different GNI and CO2 in each of the countries. It shows that most countries CO2 is increasing over the years as GNI increases from 2009 to 2018.

### Research Question 4: How have countries GNI affected their Forest Area?

```python
# GNI Vs Forest Area per country
comparison_plot(gni_clean, forest_area_clean, 'GNI', 'Forest Area', 
                'Changes in GNI against Forest Area of West African countries, 2009 - 2018.')
```

>##### The line charts above show the different GNI and forest area in each of the countries. It shows that most countries forest area is decreasing over the years as GNI increases from 2009 to 2018.

## Conclusions

#### Research Question 1: which West African countries are doing better than others in GNI, forest preservation, and controlling Co2 emissions and population?

>***GNI***:
Between the years 2009 to 2018, Nigeria takes the lead having the highest GNI, followed by Ghana, Senegal, Cote d'Ivoire and Mali in that order.

>***Forest Area***:
Nigeria also leads with the highest forest area even though its forest area is rapidly declining. Mali with the second-highest forest area has steadily preserved its forest area. Senegal with the third-highest forest area also has a declining forest area. Ghana’s forest area is slowly increasing while that of Liberia is slowly declining. 

>***CO2 emissions***:
Nigeria has the highest CO2 emissions with its highest emission recorded in 2011, there was a decline after, then a steady rise in 2015. with the second-highest emissions, Senegal’s emissions have been rising since 2012, taking over Cape Verde in 2015. As the third-highest, Cape Verde has made good progress in decreasing its emissions. Mauritania and Ghana’s emissions have also steadily increased.

>***Population***:
Nigeria is the most populated country in West Africa and its population keeps rising. Other steadily rising populations are Ghana, Cote d’Ivoire, Niger and Burkina Faso in that order.


#### Research Question 2: Can the growth in population be attributed to GNI growth in these countries?
> The line charts show that in all West African countries, the population keeps rising, even when there is a decline in GNI. For instance, in Gambia, its GNI declined several times while the population kept growing. This shows that although GNI and population are rising, GNI growth does not necessarily cause the population to grow. 

#### Research Question 3: How have West African countries' GNI affected their CO2 emissions?
> The line charts show that in all West African countries, with the exceptions of Cote d’Ivoire, Nigeria and Mauritania, CO2 emissions increases as GNI increases. Research suggests that economic activities are known to negatively impact the environment, especially during the early stages of economic growth (Jiang, M., Kim, E., & Woo, Y., 2020; Başar, S., Tosun, B., 2021). 

#### Research Question 4: How have West African countries' GNI affected their forest area?
> The line charts show in West Africa,  as the GNI increases, the forest area decreases. This exception is seen only in Cape Verde which has steadily increased its forest areas and in Mali whose forest area has remained constant in the selected years.
